```python
data.dropna()
```

```python
data['Active'] = data['Active'].fillna(data['Active'].mean())
data['Recovered'] = data['Recovered'].fillna(data['Recovered'].mean())
data['Deaths'] = data['Deaths'].fillna(data['Deaths'].mean())
```

```python
data['Active'] = data['Active'].fillna(data['Active'].median())
data['Recovered'] = data['Recovered'].fillna(data['Recovered'].median())
data['Deaths'] = data['Deaths'].fillna(data['Deaths'].median())
```

```python
data['Active'] = data['Active'].fillna(data['Active'].mode()[0])
data['Recovered'] = data['Recovered'].fillna(data['Recovered'].mode()[0])
data['Deaths'] = data['Deaths'].fillna(data['Deaths'].mode()[0])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10,6))
data.plot(kind='bar', color='red')
plt.title(" States with Highest Deaths")
plt.xlabel("State")
plt.ylabel("Total Deaths")
plt.xticks(rotation=95)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.bar(data['Name of State / UT'], data['Active'], color='skyblue')
plt.xlabel("Name of State / UT")
plt.ylabel("Active Cases")
plt.title("Active COVID-19 Cases by State/UT")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
